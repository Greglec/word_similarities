```python
#basic libraries
import pandas as pd
import numpy as np
import datetime

#library to compute % similarity between strings
from difflib import SequenceMatcher
from difflib import get_close_matches
```

```python
odoo_leads = pd.read_excel('C:/Users/grego/Desktop/i-lunch/Marketing/CRM/leads/leads_odoo.xlsx')

#on retire les lignes en double
odoo_leads = odoo_leads.drop_duplicates()

odoo_leads.fillna("Nan",inplace=True)
```

```python
new_leads = pd.read_excel('C:/Users/grego/Desktop/i-lunch/Marketing/CRM/leads/Pistes Global - All in one sheet.xlsx')
new_leads.fillna("Nan",inplace=True)
```

```python
#tratiement des numeros de tel

def right(s, amount):
    return s[-amount:]

#creation de id phone pour check 4
new_leads["id_tel"] = new_leads["Téléphone"].str.replace(" ","").astype(str)
odoo_leads["id_tel"] = odoo_leads["Téléphone"].str.replace(" ","").astype(str)
odoo_leads["id_tel2"] = odoo_leads["Mobile"].str.replace(" ","").astype(str)
new_leads["id_tel"] = new_leads.apply(lambda x: right(x['id_tel'],8),axis =1)
odoo_leads["id_tel"] = odoo_leads.apply(lambda x: right(x['id_tel'],8),axis =1)
odoo_leads["id_tel2"] = odoo_leads.apply(lambda x: right(x['id_tel2'],8),axis =1)
```

```python
########################################### check 1 #######################################
list1 = list(map(lambda x:x.lower(),odoo_leads['Nom de la société']))
sim_company_name =  pd.DataFrame(pd.np.empty((len(new_leads), 2))) 
sim_company_name.columns = ["Entreprise","Check1"]
sim_company_name["Check1"] = sim_company_name["Check1"].astype(str)
sim_company_name["Entreprise"] = new_leads["Entreprise"]
sim_company_name["Entreprise"] = sim_company_name["Entreprise"].astype(str)

for i in range(0,sim_company_name.shape[0]):
    print(round((i/len(new_leads)),2) *100, "%")
    try:
        sim_company_name.iat[i,1] = get_close_matches(list(map(lambda x:x.lower(),sim_company_name["Entreprise"]))[i], list(map(lambda x:x.lower(),list1)), 1, 0.8)[0]
    except:
        sim_company_name.iat[i,1] = "Pas de correspondance"
```

```python
########################################### check 2 #######################################
sim_mail =  pd.DataFrame(pd.np.empty((len(new_leads), 2))) 
sim_mail.columns = ["Email","Check2"]
sim_mail["Check2"] = sim_mail["Check2"].astype(str)
sim_mail["Email"] = new_leads["Email"]
sim_mail["Email"] = sim_mail["Email"].astype(str)

list1 = list(map(lambda x:x.lower(),odoo_leads['Email']))
for i in range(0,sim_mail.shape[0]):
    print(round((i/len(new_leads)),2) *100, "%")
    try:
        sim_mail.iat[i,1] = get_close_matches(list(map(lambda x:x.lower(),sim_mail["Email"]))[i], list(map(lambda x:x.lower(),list1)), 1, 0.9)[0]
    except:
        sim_mail.iat[i,1] = "Pas de correspondance"
```

```python
sim_mail["Entreprise"] = new_leads["Entreprise"]
sim_mail["Entreprise"] = sim_mail["Entreprise"].astype(str)
```

```python
########################################### check 3 #######################################

sim_phone =  pd.DataFrame(pd.np.empty((len(odoo_leads), len(new_leads)))) 
sim_phone.columns = new_leads["id_tel"]
sim_phone.index = odoo_leads['id_tel']

sim_phone =  pd.DataFrame(pd.np.empty((len(new_leads), 2))) 
sim_phone.columns = ["id_tel","Check3"]
sim_phone["Check3"] = sim_phone["Check3"].astype(str)
sim_phone["id_tel"] = new_leads["id_tel"]
sim_phone["id_tel"] = sim_phone["id_tel"].astype(str)

list1 = odoo_leads['id_tel']
for i in range(0,sim_phone.shape[0]):
    if round((i/len(new_leads)),2) *100 > round(((i-1)/len(new_leads)),2) *100:
        print(round((i/len(new_leads)),2) *100, "%")
    try:
        sim_phone.iat[i,1] = get_close_matches(sim_phone["id_tel"][i], list1, 1, 0.9)[0]
    except:
        sim_phone.iat[i,1] = "Pas de correspondance"
```

```python
sim_phone["Entreprise"] = new_leads["Entreprise"]
sim_phone["Entreprise"] = sim_phone["Entreprise"].astype(str)
```

```python
########################################### check 4 #######################################
sim_name =  pd.DataFrame(pd.np.empty((len(new_leads), 2))) 
sim_name.columns = ["Name","Check4"]
sim_name["Check4"] = sim_name["Check4"].astype(str)
sim_name["Name"] = new_leads["Nom"]
sim_name["Name"] = sim_name["Name"].astype(str)

list1 = list(map(lambda x:x.lower(),odoo_leads['Nom']))
for i in range(0,sim_name.shape[0]):
    print(round((i/len(new_leads)),2) *100, "%")
    try:
        sim_name.iat[i,1] = get_close_matches(list(map(lambda x:x.lower(),sim_name["Name"]))[i], list(map(lambda x:x.lower(),list1)), 1, 0.9)[0]
    except:
        sim_name.iat[i,1] = "Pas de correspondance"
        
sim_name["Entreprise"] = new_leads["Entreprise"]
sim_name["Entreprise"] = sim_name["Entreprise"].astype(str)
```

```python
def split(my_string, split_string):
    try:
        result = my_string.split(split_string,1)[1] 
    except:
        result = my_string.split(split_string,1)[0] 
    return result
```

```python
#creation de id site web pour check 6 & 7
new_leads["id_web"] = new_leads["Site web de l'entreprise"].str.replace(" ","").astype(str)
odoo_leads["id_web"] = odoo_leads["Site Web"].str.replace(" ","").astype(str)
new_leads["id_web"] = new_leads.apply(lambda x: split(x['id_web'],"//"),axis =1)
odoo_leads["id_web"] = odoo_leads.apply(lambda x: split(x['id_web'],"//"),axis =1)
```

```python
########################################### check 6 #######################################
sim_web =  pd.DataFrame(pd.np.empty((len(new_leads), 2))) 
sim_web.columns = ["id_web","Check6"]
sim_web["Check6"] = sim_web["Check6"].astype(str)
sim_web["id_web"] = new_leads["id_web"]
sim_web["id_web"] = sim_web["id_web"].astype(str)

list1 = list(map(lambda x:x.lower(),odoo_leads['id_web']))
for i in range(0,sim_web.shape[0]):
    if round((i/len(new_leads)),2) *100 > round(((i-1)/len(new_leads)),2) *100:
        print(round((i/len(new_leads)),2) *100, "%")
    try:
        sim_web.iat[i,1] = get_close_matches(list(map(lambda x:x.lower(),sim_web["id_web"]))[i], list(map(lambda x:x.lower(),list1)), 1, 0.7)[0]
    except:
        sim_web.iat[i,1] = "Pas de correspondance"
        
sim_web["Entreprise"] = new_leads["Entreprise"]
sim_web["Entreprise"] = sim_web["Entreprise"].astype(str)
```
